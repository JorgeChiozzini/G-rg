```python
# Introdução
# Este notebook implementa a detecção de bordas em imagens usando o filtro de Sobel

# Importando bibliotecas necessárias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregar a imagem
image = cv2.imread('/workspaces/G-rg/edge-detection/girassol.jpg.webp', cv2.IMREAD_GRAYSCALE)

# Verificar se a imagem foi carregada corretamente
if image is None:
    print("Erro ao carregar a imagem.")
else:
    # Exibir a imagem original
    plt.figure(figsize=(8, 8))
    plt.imshow(image, cmap='gray')
    plt.title('Imagem Original')
    plt.axis('off')
    plt.show()
```

```python
# Aplicar o filtro de Sobel para detectar bordas
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Calcular a magnitude das bordas
sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
sobel_magnitude = np.uint8(sobel_magnitude / np.max(sobel_magnitude) * 255)

# Exibir a imagem resultante
plt.figure(figsize=(8, 8))
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Detecção de Bordas com Sobel')
plt.axis('off')
plt.show()
```
